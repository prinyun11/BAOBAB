```python
import pandas as pd
import numpy as np
```

```python
# 파일 불러오기

salon = pd.read_csv('/Users/annakim/Library/CloudStorage/OneDrive-성신여자대학교/BOAZ/ADV/버블클라우드/data/salon_salon.csv')
booking = pd.read_csv('/Users/annakim/Library/CloudStorage/OneDrive-성신여자대학교/BOAZ/ADV/버블클라우드/data/booking_booking.csv')
```

```python
# 파일 확인

salon.head()
```

```python
booking.head()
```

```python
# 겹치는 컬럼 이름 바꾸기

salon.rename(columns = {'id' : 'salon_id', 'created_at' : 'created_at_salon'}, inplace = True)
salon.head()
```

```python
booking.rename(columns = {'created_at' : 'created_at_booking'}, inplace = True)
booking.head()
```

```python
# salon, booking 데이터프레임 합치기

data = pd.merge(salon, booking, on = 'salon_id')
data.head()
```

```python
# created_at_booking에서 날짜만 뽑아서 created_date 컬럼에 넣어주기

temp = []
for i in range(data.shape[0]):
    temp.append(str(data['created_at_booking'][i]).split(' ')[0])
    
data['created_date'] = temp
data.head()
```

```python
# 날짜 기준으로 묶기

df = pd.DataFrame(data.groupby('created_date')[['salon_id', 'name']])
df.rename(columns = {0 : 'date'}, inplace = True)
df
```

```python
# 확인 일자 기준 7일 이내 생성된 예약만 확인

temp = df[(df['date'] <= '2022-05-20') & (df['date'] >= '2022-05-14')]
temp.reset_index(inplace = True)
temp
```

```python
# 미용실별 해당 날짜에 생성된 예약의 수를 구해 하나의 데이터프레임으로 만들고
# 예약이 10건 이상인 것들만 활성화 샵으로 뽑기

df1 = pd.DataFrame(temp[1][0].value_counts()).reset_index().rename(columns = {0 : 'count'})
df2 = pd.DataFrame(temp[1][1].value_counts()).reset_index().rename(columns = {0 : 'count'})
df3 = pd.DataFrame(temp[1][2].value_counts()).reset_index().rename(columns = {0 : 'count'})
df4 = pd.DataFrame(temp[1][3].value_counts()).reset_index().rename(columns = {0 : 'count'})
df5 = pd.DataFrame(temp[1][4].value_counts()).reset_index().rename(columns = {0 : 'count'})
df6 = pd.DataFrame(temp[1][5].value_counts()).reset_index().rename(columns = {0 : 'count'})
df7 = pd.DataFrame(temp[1][6].value_counts()).reset_index().rename(columns = {0 : 'count'})

active = pd.concat([df1, df2, df3, df4, df5, df6, df7]).groupby(['salon_id', 'name'])['count'].sum().sort_values(ascending = False)
active = pd.DataFrame(active)
active = active[active['count'] >= 10]
active
```

```python
# 파일로 저장

active.to_csv('active_salon.csv')
```
