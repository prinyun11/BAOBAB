```python
import pandas as pd
from datetime import datetime
import time
```

```python
# 파일 불러오기

active = pd.read_csv('/Users/annakim/Library/CloudStorage/OneDrive-성신여자대학교/BOAZ/ADV/버블클라우드/data/활성샵_데이터/active_salon.csv')
active.drop('name', axis = 1, inplace = True)
active.head()
```

```python
booking = pd.read_csv('/Users/annakim/Library/CloudStorage/OneDrive-성신여자대학교/BOAZ/ADV/버블클라우드/data/booking_booking.csv')
booking.head()
```

```python
# 활성샵 id만 리스트로 추출

active_salon_id = list(active['salon_id'])
```

```python
# 활성샵 id에 해당하는 booking만 추출

df = pd.DataFrame()
for i in active_salon_id:
    temp = booking[booking['salon_id'] == i]
    df = pd.concat([df, temp])
```

```python
# 필요 없는 컬럼 삭제

df.drop(['id', 'check_out', 'dog_id', 'groomer_id', 'color_id', 'is_active', 'album_id', 'created_at'], axis = 1, inplace = True)
df
```

```python
# 인덱스 번호 초기화

df.reset_index(inplace = True)
df.head()
```

```python
# date 컬럼 타입 확인

df['date'].info()
```

```python
# date 컬럼 타입 datetime으로 바꾸기

date = list(df['date'])
temp = []
for i in date:
    datetime_string = i
    datetime_format = "%Y-%m-%d"

    temp.append(datetime.strptime(datetime_string, datetime_format))

df['date'] = temp
df.head()
```

```python
# 요일 구하기

days = {0: '월요일', 1: '화요일', 2: '수요일', 3: '목요일', 4: '금요일', 5: '토요일', 6: '일요일'}
temp = []

for i in range(df.shape[0]):
    temp.append(days[df['date'][i].weekday()])
    
df['day'] = temp
df.head()
```

```python
# 파일 내보내기

df.to_csv('liked_dates.csv')
```
